# Stemming

Stemming is the process of reducing a word to its root/base form by removing prefixes or suffixes, so that different variations of the same word are converted into a common form.

It is a text preprocessing technique used in Natural Language Processing (NLP).

```python
## classification problem
## comments of product is a positive review or negative review.
## reviews ----->> eating, eat, eaten [going, gone, goes] ---> go

```

## PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
words = [
    "eating",
    "eats",
    "eaten",
    "writing",
    "writes",
    "programming",
    "programs",
    "history",
    "finally",
    "finalized"
]
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+"---->" + stemming.stem(word))
```

```python
## disadvantage of stem -- not giving correct word
stemming.stem('Congratulations')
```

# RegexpStemmer class
RegexpStemmer is a stemming class in NLTK that removes a specific pattern (regular expression) from words to create their stem.

Unlike PorterStemmer, which uses predefined English rules, RegexpStemmer lets you define your own rule using regex.

```python
from nltk.stem import RegexpStemmer
```

```python
# it is saying wherever last word is ing starting with s, e, able....just remove it 

reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min = 4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

#Snowball Stemmer

Snowball Stemmer is an improved version of the Porter Stemmer algorithm used for stemming words. It is a rule-based stemming algorithm that reduces words to their root/base form by removing suffixes.

It is called Snowball because the algorithm was developed in the Snowball programming language and supports multiple languages.


```python
from nltk.stem import SnowballStemmer
```

```python
snowballsstemeer = SnowballStemmer('english')
```

```python
for word in words:
     print(word + "--->" + snowballsstemeer.stem(word))
```
