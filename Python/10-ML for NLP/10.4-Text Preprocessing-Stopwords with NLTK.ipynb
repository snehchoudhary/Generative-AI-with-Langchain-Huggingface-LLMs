```python
## speech of APJ Abdul Kalam

paragraph = """Dream, Dream, Dream — Speech style

My dear friends,

Dream is not what you see while sleeping; dream is something that does not let you sleep. Every great achievement begins with a dream, a vision, and a strong determination to make it a reality.

When you have a dream, you must work hard with confidence and dedication. Difficulties will come in your path, but challenges are not meant to stop you; they are meant to make you stronger.

You should always remember that your future depends on what you do today. Have a great aim, continuously acquire knowledge, work hard, and never give up.

Small goals do not inspire people. Great dreams create great minds. Believe in yourself and believe that you have the power to create change.

The youth of today are the leaders of tomorrow. Use your knowledge, creativity, and energy to build a better future for yourself and for the nation.

Always keep learning, always keep improving, and always keep moving forward.

Thank you."""
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk 
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
stopwords.words('german')
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
##apply stopwords and filter and then apply snowball stemming

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]

    sentences[i] = ''.join(words)   #converting all the words into sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```
