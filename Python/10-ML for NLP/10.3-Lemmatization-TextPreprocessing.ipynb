# Wordnet Lemmatizer

WordNet Lemmatizer is a lemmatization technique in NLTK that converts words into their base dictionary form (lemma) using the WordNet lexical database.

Unlike stemming, lemmatization does not just remove suffixes. It understands the meaning and context of words and returns a valid English word.

Lemmatization technique is like stemming. The output we will get after lemmatization is called 'lemma', which is a root word rather than root stem, the output of stemming. after lemmatization, we will be getting a validword that means the same thing.

```python
##use cases - Q&A, cahtbots, text, summarization

import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
''' 
POS -Noun -n
verb -v
adjective-a
adverb-r
'''

lemmatizer.lemmatize("going", pos='v')
```

```python
words = [
    "eating",
    "eats",
    "eaten",
    "writing",
    "writes",
    "programming",
    "programs",
    "history",
    "finally",
    "finalized"
]
```

```python
for word in words:
    print(word + "---->" + lemmatizer.lemmatize(word, pos='v'))
```
