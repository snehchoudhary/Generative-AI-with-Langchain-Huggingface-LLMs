The map() Function in Python

The map() function applies a given function to all items in an input list(or any other iterable) and return a map object(an iterator). This is particularly useful for transforming data in a list comprehensively.

```python
def square(x):
    return x*x
square (10)
```

```python
numbers=[1, 2, 3, 4,5, 6,7,8]
list(map(square, numbers))
```

```python
## lambda function with map
numbers=[1,2,3 ,4 ,5 ,6,7, 8]
list(map(lambda x : x*x, numbers))
```

```python
##map multiple iterables

numbers1=[1,2,3]
numbers2 = [4,5,6]

added_numbers = list(map(lambda x, y: x + y, numbers1, numbers2))

print(added_numbers)
```

```python
## map() to convert a list strings to integers
# use map to convert strings to integers

str_numbers = ['1', '2', '3', '4', '5']
int_numbers = list(map(int, str_numbers))

print(int_numbers)  
```

```python
words = ['apple', 'banana', 'cherry']
upper_words = list(map(str.upper, words))
print(upper_words)
```

```python
def get_name(person):
    return person['name']

people = [
    {'name' : 'Sneha', 'age' : 20},
    {'name': 'Jack', 'age' : 22}
]
list(map(get_name, people))
```
