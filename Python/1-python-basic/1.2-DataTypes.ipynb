## DataTypes

1. Definition:
 
Data types are a classification of data which tell the compiler or interpreter how the programmer intends to use the data.
 
 They dtermine the type of operation that can be performed on the data, the values that the data can take, and the amount of emory needed to store the data.

2. Importance of Data Types in Programming

Explanation:

Data types ensure that data is stored in an efficient way.

They help in performing correct operations on data.

Proper use of Data types can prevent errors and bugs in the program.

```python
##Integer
age = 35
type(age)
```

```python
## floating point datatype
height = 5.0
print(height)
print(type(height))
```

```python
## string data type
name = "Sneha"
print(type(name))
```

```python
# boolean data type
is_true = True
type(is_true)
```

```python
## error
result = "hello" + 5
```

```python
#fix the error
result = "hello" + str(5)
print(result)
```
