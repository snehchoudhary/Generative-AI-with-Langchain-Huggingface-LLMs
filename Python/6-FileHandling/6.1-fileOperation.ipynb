File operation - Read and Write Files

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and bianry. This lesson will cover the basics of file handling, including reading and writing text files and binary files.

```python
## read a whole file

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
## read a file line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())   #sstrip() removes the newline charater
```

```python
## writing a file(overwriting)

with open('example.txt', 'w') as file:
    file.write('hello world!\n')
    file.write('this is a new line.')
```

```python
#write a file(without overwriting)

with open('example.txt', 'a') as file:
    file.write("Append operation taking place!")
```

```python
## writing a list of lines to a file
lines = ['First line \n', 'second line \n', 'hird line \n']
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
## binary files

# writing to a bianry file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
# reading a bianry file

with open ('example.bin', 'rb') as file:
    content = file.read()
    print(content) 
```

```python
# read the content  from source text file and write to a destiantion text file

#copying a text file
with open('example.txt', 'r') as  source_file:
 content = source_file.read()

with open('destination.txt', 'w') as destination_file:
 destination_file.write(content)

```
