## Introduction to Tuples
Tuples are ordered collection of items that are immutable. They are similar to lists, but their immutability makes them different


```python
# create a tuple
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers = tuple([1,2 , 3, 4, 5, 6])
numbers
```

```python
list((1,2, 3, 4, 5, 6))
```

```python
mixed_tuple = (1, "hello world", 3.14, True)
print(mixed_tuple)
```

```python
# acessing tuple elements
numbers
```

```python
print(numbers[2])
print(numbers[-1])
```

```python
numbers[0:4]
```

```python
numbers[::-1]
```

```python
# tuple operations

concatenation_tuple = numbers + mixed_tuple
print(concatenation_tuple)
```

```python
mixed_tuple * 3
```

```python
numbers * 3
```

```python
## immutable nature of tuples
## tuples are immutable, means their elements caanot be changed once assigned.

lst=[1, 2, 3, 4, 5]
print(lst)

lst[1] = "Sneha"
print(lst)
```

```python
numbers[1] = "Sneha"
```

```python
numbers
```

```python
#tuple methods

print(numbers.count(1))
print(numbers.index(3))
```

```python
## packing and unpacking tuple
## packing
packed_tuple = 1, "hello", 3.14
print(packed_tuple)
```

```python
#unpacking a tuple
a, b, c = packed_tuple
print(a)
print(b)
print(c)
```

```python
# unpacking with *
numbers = (1, 2, 3, 4, 5, 6)
first, *middle, last = numbers
print(first)
print(middle)
print(last)
```

```python
# nested tuple

#nested list
lst = [[1, 2, 3, 4], [5, 6, 7, 8], [1, "hello", 3.14, "c"]]
lst[0][2]
```

```python
lst = [[1, 2, 3, 4], [5, 6, 7, 8], [1, "hello", 3.14, "c"]]
lst[2][0:3]
```

```python
nested_tuple = ((1,2,3), ("a", "b", "c"), (True, False))

## acess the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
# iterating over nested tuple
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=" ")
    print()
```
