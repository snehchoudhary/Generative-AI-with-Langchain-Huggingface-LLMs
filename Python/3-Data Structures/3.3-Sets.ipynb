## Introduction to Dictionaries

Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable(e.g., strings, numbers, or tuples), whilevalues can be of any type.

```python
## creating dictionaries
empty_dictionary={}
print(type (empty_dictionary))

```

```python
empty_dictionary=dict()
empty_dictionary
```

```python
student = {"name": "Sneha", "age": 20, "grade": 87}
print(student)
print(type(student))
```

```python
#single keys is always used
student = {"name": "Sneha", "age" : 20, "name" : 45}
print(student)
```

```python
# acessing dict elements
student = {"name": "Sneha", "age": 20, "grade": 'A'}
print(student)
```

```python
print(student['grade'])
print(student['age'])

## acessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name', "not available"))
```

```python
## modifying dict elements
## dict are mutable so we can add, update or delete elements
print(student)
```

```python
student["age"] = 22    ## updated the value for the key
print(student)
student["address"] = "India"   ## added a new key and value
print(student)
```

```python
del student['grade'] ## delete key and value pair
print(student)
```

```python
## dictionary methods

keys = student.keys()    ##get all the keys
print(keys)
values = student.values()  ## get all values
print(values)
items = student.items()    ##get all key value pairs
print(items)
```

```python
## shallow copy
student_copy = student
print(student)
print(student_copy)
```

```python
student["name"] = "Sneha2"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy()   # shallow copy
print(student_copy1)
print(student)
```

```python
student["name"] = "Sneha3"
print(student_copy1)
print(student)
```

```python
## iterating over dictionaries
# we can use loops to iterate over dictionaries, keys, values, or items

#iterating over keys
for keys in student.keys():
    print(keys)
```

```python
# iterate over values
for value in student.values():
    print(value)
```

```python
# iterate over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## nested dictionaries
students= {
    "student1":{"name":"Sneha", "age": 20}, 
    "student2":{"name":"peter", "age" : 34}
}
print(students)
```

```python
#acess nested dict
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key, value in student_info.items():
        print(f"{key}:{value}")

```

```python
## dictionary comprehension
squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
## conditional comprehension
evens = {x:x**2 for x in range(10) if x%2 == 0}
print(evens)
```

```python
## use a dict to count the frequency of elements in list

numbers=[1, 2,2, 3, 3, 3, 4,4, 4, 4]
frequency ={}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
print(frequency)
```

```python
#merge 2 dictionaries
dict1 = {"a" : 1, "b":2}
dict2 = {"b" : 3, "c" : 4}
merged_dict = {**dict1, **dict2}
print(merged_dict)
```
