Standard Library Overview

Python's Standard Library is a vast collection of modules and pacakges that comebundled with python, providing a wide range of functionalities out of the box. Here's an  overview of some of the most commonly used modules and packages in the Python Standard Library.

```python
import array
arr = array.array('i', [1,2,3,4])
print(arr)
```

```python
import math
print(math.sqrt(16))
print(math.pi)
```

```python
## random

import random
print(random.randint(1, 10))
print(random.choice(['apple', 'banana', 'cherry']))
```

```python
# high level operations on files and collections of files
import shutil
shutil.copyfile('source.txt', 'destination.txt')
```

```python
# data serialization
import json 
data = {'name': "Sneha", 'age': 20}

json_str = json.dumps(data)
print(json_str)
print(type(json_str))

parsed_data=json.loads(json_str)
print(parsed_data)
print(type(parsed_data))

```

```python
## csv

import csv
with open('example.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['name', 'age'])
    writer.writerow(['Sneha', 20])

with open('example.csv', mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            print(row)
```

```python
## datetime
from datetime import datetime, timedelta

now=datetime.now()
print(now)

yesterday = now-timedelta(days=1)

print(yesterday)
```

```python
## time
import time
print(time.time())
time.sleep(2)
print(time.time())
```

```python
# regular expression
import re

pattern=r'\d+'
text = 'there are 123 apple'
match= re.search(pattern, text)
print(match.group())
```
